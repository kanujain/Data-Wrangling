```python
df1 = {"s_id": [1,2,3,4,5],
       "s_month": ['Jan','feb','mar', 'apl', 'may'],
       "sales": [100,120,321,123,543]}
import pandas as pd

df1 = pd.DataFrame(df1, columns = ["s_id","s_month","sales"])
df1
```

```python
df2 = {"s_id": [6,7,8,9,10],
       "s_month": ['june','july','aug', 'sep', 'oct'],
       "sales": [51516,6546515,65465,652,221]}
df2 = pd.DataFrame(df2, columns = ["s_id","s_month","sales"])
df2
```

```python
df3 = {'s_threshold': ['n','y','n','y','y','y','y','y','y','y'],
       'bonus_threshold': ['n','y','n','n','n','y','n','y','y','n']}

df3 = pd.DataFrame(df3, columns =['s_threshold','bonus_threshold'])
df3
```

```python
#concat() : when we want to add rows

df4 = pd.concat([df1,df2], ignore_index = True) #ignore_index avoid using the idexes of previous dataset
df4
```

```python
df5 = pd.concat([df4,df3], axis=1) #adding column
df5
```

```python
df6 = df1.append(df2)  #more useful in adding rows not columns

df6
```

```python
df7 = df1.append(df3)
df7
```
