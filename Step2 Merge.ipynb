```python
Merge: 1-1, 1-many, many to many, many-1
```

```python
import pandas as pd
```

```python
df1 = {"s_month": ['Jan','feb','mar', 'apl', 'may'],
       "product": ['a','b','c','d','e']}

df1 = pd.DataFrame(df1, columns = ["s_month","product"])

df2 = {"s_month": [ 'may','Jan','mar', 'feb','apl'],
       "sales": [51516,6546515,65465,652,221]}
df2 = pd.DataFrame(df2, columns = ["s_month","sales"])

display(df1,df2)
```

```python
df3 = pd.merge(df1,df2)
df3
```

```python
d1 = {'Product': ['p1','p2','p3','p4'],
     'Division': ['a','b','c','d']}
d2 = {'sales': [10,20,30,40],
     'Product': ['p3','p2','p1','p4']}
d1 = pd.DataFrame(d1, columns = ["Product","Division"])
d2 = pd.DataFrame(d2, columns = ["sales","Product"])

display(d1,d2)
```

```python
display(pd.merge(d1,d2,on = 'Product'))  # wrt primary key
```

```python
d3 = {'Project': ['p1','p2','p3','p4'],
     'Division': ['a','b','c','b']}
d3 = pd.DataFrame(d3, columns = ["Project","Division"])
d3
```

```python
#left_on, right_on

display(pd.merge(d2,d3,left_on='Product', right_on='Project').drop('Project', axis=1))
```

```python
df1 = d1.set_index('Product')
df2 = d2.set_index('Product')

display(df1,df2)
```

```python
display(pd.merge(df1,df2, left_index = True, right_index= True))
```

```python
#Joins
    Inner
    outer
    left
    right
    
    
```

```python
pd.merge(d1,d2, how='left')
pd.merge(d1,d2, how='inner')
pd.merge(d1,d2, how='right')
pd.merge(d1,d2, how='outer', indicator=True)  # from which amongst above 3 does the query combine
```
