```python
Stack and Unstack

    # stack : wide to long
    #unstack : long to wide

Used when the data_set is multi_Scaled
    #eg: salary of person before and after i.e. same feature but different aspect 

```

```python
#Creating multi Indexed data frame

header = pd.MultiIndex.from_product([['Before_upskilling', 'After_Up_skilling'], ['Age','Salary']])
data = [ [25,40000, 28, 80000], [38,5000000, 49, 3240000000], [21, 4200000000, 32, 5600000000]]

data_set = pd.DataFrame(data,
                       index = ['A','B','C'],
                       columns = header)
data_set
```

```python
data_stacked = data_set.stack()  #By default level = 1 i.e. which index we want to use as column
data_stacked
```

```python
data_set.stack(level=0)  # with level =0 we marked after and before row as column

        #Level depends on the level of multi data_set we created/ have
```

```python
#Unstacking: long to wide

data_stacked.unstack()
```

```python
#PIVOT TABLE
    By default it aggregates along its index
    works with numeric data and drops non-numeric
    mean is default aggregate function, can be changed
```

```python
data = pd.read_csv("E:\\2020\\Quarantine\\Python\\prices.csv")
data.head()
```

```python
stock_pivottable = pd.pivot_table(data, index='symbol')  #we have to assign index to a pivot table
stock_pivottable.head()
```

```python
p= data.groupby('symbol')['close'].mean()
p.head()

g= data.groupby('symbol')['close'].sum()
g.head()
```

```python
pd.pivot_table(data, index=['symbol', 'date'])   #Including multiple indexes
```

```python
import numpy as np

pd.pivot_table(data, index='symbol', aggfunc=np.sum, values='volume')  #applying what function we want to work with
```
